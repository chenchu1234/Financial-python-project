```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv(r"C:\Users\user\Desktop\krishna1@\New folder\Financial_Data_Analysis 123.csv")
df
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df=df.fillna(0)
df
```

```python
# Define a NUll Values
df.isnull().sum()
```

```python
# Convert Transaction_Date into DateTime Type
df['Transaction_Date'] = pd.to_datetime(df['Transaction_Date'], errors='coerce')
print(df['Transaction_Date'])
```

```python
# Find the total number of customers by Gender
df['Gender'].value_counts()
```

```python
# what is the average transaction Amount
df['Amount'].mean()
```

```python
# Which region has the highest number of customers
df['Region'].value_counts()
```

```python
# Count how many loans were approved vs rejected
df['Loan_Status'].value_counts()
```

```python
# Find the Average credit score by region
df.groupby('Region')['Credit_Score'].mean()
```

```python
# How many transaction of each occurred
df['Transaction_Type'].value_counts()
```

```python
# Find the minimum, maximum, and average age
df['Age'].describe()
```

```python
# Plot number of customers by region (Bar Charts)
import matplotlib.pyplot as plt

df['Region'].value_counts().plot(kind='bar')
plt.title('Customer by Region')
plt.xlabel('Region')
plt.ylabel('Count')
plt.show()
```

```python
# Plot Credit Score distribution

plt.hist(df['Credit_Score'], bins=20)
plt.title('credit Score Distribution')
plt.xlabel('Credit_Score')
plt.ylabel('Frequency')
plt.show()
```

```python
# Identify customers with Credit Score < 400 (High Risk Customers)

high_risk = df[df['Credit_Score'] < 400]
high_risk.head()
```

```python
# Calculate total transaction amount for each transaction type
df.groupby('Transaction_Type')['Amount'].sum()
```

```python
# Find regions with highest loan approval rate 

approval_rate = (
    df[df['Loan_Status']=='Approved']
    .groupby('Region')
    .size()
)
approval_rate
```

```python
df['Year'] = df['Transaction_Date'].dt.year
df.head()
```

```python
# Total number of transaction per year
df.groupby('Year').size()
```
